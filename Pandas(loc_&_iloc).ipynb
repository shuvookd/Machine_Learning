```python
import pandas as pd
```

```python
df = pd.read_csv('/content/student_results.csv')
df
```

### **loc()**

```python
#Print First row values
df.loc[0]
```

```python
#Print 4th row values
df.loc[4]
```

```python
#Print specific multiple row values
df.loc[[2,4]]
```

```python
#Print specific column values
df.loc[4, 'Class']
```

```python
#Print multiple specific columns values
df.loc[0:3, 'Percantege']
```

```python
#Use conditional Operator
df.loc[df['Class']<11]
```

```python
#Use conditional Operator
df.loc[df['Class']<11, ['Percantege']]
```

### **iloc()**

```python
#Print First Column
df.iloc[[0]]
```

```python
#Print all rows, First Column
df.iloc[:, 0]
```

```python
#Print specific rows
df.iloc[[5,3,7]]
```
