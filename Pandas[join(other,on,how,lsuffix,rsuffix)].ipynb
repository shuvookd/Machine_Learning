### **join()**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B': [10,20,30]},index = ['a','b','c'])
df2 = pd.DataFrame({'C': [4,5,6],
                    'D': [40,50,60]},index = ['a','b','c'])
display(df1,df2)
```

```python
#Join 2 Dataframe
df1.join(df2)
```

```python
df1.join(df2, how = 'right') #left,righr,inner,outer
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B': [10,20,30]},index = ['a','b','c'])
df2 = pd.DataFrame({'A': [4,5,6],
                    'B': [40,50,60]},index = ['a','b','c'])
display(df1,df2)
```

```python
#if colomns names are same use lsuffix,rsuffix
df1.join(df2, lsuffix  = '_1')
```
