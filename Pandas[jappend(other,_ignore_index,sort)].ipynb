### **append()**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B': [10,20,30]})
df2 = pd.DataFrame({'A': [4,5,6],
                    'B': [40,50,60]})
display(df1,df2)
```

```python
#To correct Index
df1.append(df2, ignore_index = True)
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B': [10,20,30]})
df2 = pd.DataFrame({'C': [4,5,6],
                    'B': [40,50,60]})
display(df1,df2)
```

```python
#To Sort
df1.append(df2, sort = False)
```
