### **Changing unnecessary values to NaN**

```python
df1 = pd.read_csv('/content/Fortune_10.csv', na_values = ['Health'])
df1
```

### **Changing specific column's specific values to NaN**

```python
df2 = pd.read_csv('/content/Fortune_10.csv', na_values = {'Name': 'Lamtone'})
df2
```
