### Finding missing values or null Values by **.isnull()**

```python
#Finding null values exists or not
df.isnull()
```

```python
#To know how much null values in a column
df.isnull().sum()
```

```python
#To know total null values
df.isnull().sum().sum()
```

### Finding missing values or null Values by **.notnull()**

```python
#Finding true values exists or not
df.notnull()
```

```python
#To know how much true values in a column
df.notnull().sum()
```

```python
#To know total true values
df.notnull().sum().sum()
```

### Drop a column which contains null values

**.dropna()**

```python
df.dropna()
```

```python
#Axis = 1 (Drops those columns who contains NaN Values)
df.dropna(axis = 1)
```

```python
#how = 'any' (Drops those rows who contains NaN Values)[axis = 0] by default
df.dropna(how = 'any')
```

```python
#how = 'any' (Drops those columns who contains NaN Values)
df.dropna(axis = 1, how = 'any')
```

```python
#how = 'all' (Drops those row who contains all NaN Values)
df.dropna(how = 'all')
```

```python
#thresh =1 (Drops those row who contains all NaN Values, remain thos rows who have at least 1 null value)
df.dropna(thresh = 1)
```

```python
#subset = ["ID"] (Drops those row who contains  NaN Values)
df.dropna(subset = ['ID'])
```

```python
#inplace = True (Drops those row who contains all NaN Values)
df.dropna(inplace = True)
```
