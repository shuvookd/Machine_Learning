### **To read specific first 2 Rows**

```python
df1 = pd.read_csv('/content/Fortune_10.csv', nrows = 2)
df1
```

### **To read specific Single/Multiple Columns**

```python
df2 = pd.read_csv('/content/Fortune_10.csv', usecols= [2])
df2
```

```python
df3 = pd.read_csv('/content/Fortune_10.csv', usecols= [2,4,7])
df3
```

### **To Skip Specific single/Multiple Rows by adding []**

```python
df4 = pd.read_csv('/content/Fortune_10.csv', skiprows= 0)
df4
```

```python
df5 = pd.read_csv('/content/Fortune_10.csv', skiprows= [1,2,3])
df5
```

### **Making Any column as index**

```python
#trick 1
df6 = pd.read_csv('/content/Fortune_10.csv', index_col = ['ID'])
df6
```

```python
##trick 2
df7 = pd.read_csv('/content/Fortune_10.csv', index_col = [2])
df7
```

### **Get any Column as Header 0,1,2...n**

```python
df8 = pd.read_csv('/content/Fortune_10.csv', header = 1)
df8
```

### Remove header by  **header = None**

```python
df9 = pd.read_csv('/content/Fortune_10.csv', header = None)
df9
```

### Make a sequence header by addind **prefix = 'name'**

```python
df10 = pd.read_csv('/content/Fortune_10.csv', header = None, prefix = 'Data')
df10
```

### **Add specific column name**

```python
df11 = pd.read_csv('/content/Fortune_10.csv', names = ['ID','Names', 'Industry','Inception', 'Revenue', 'Expenses', 'Profit', 'Growth'])
df11
```
